```python
#1
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, sin, diff
from sympy.utilities.lambdify import lambdify
# Define x(t)
def f(t,a): 
    return 1/2*a*t**2


a,x=symbols('a x', real=True)
a=2
print(diff(1/2*a*x**2,x))

# Define parabola derivative
def slope(x): 
    return 1.0*a*x

# Define x data range for parabola
x = np.linspace(0,5,100)

# Choose point to plot tangent line
x1 = 1.5
y1 = f(x1,a)

# Define tangent line
# y = m*(x - x1) + y1
def line(x, x1, y1):
    return slope(x1)*(x - x1) + y1

# Define x data range for tangent line
xrange = np.linspace(x1, x1+2, 10)

# Plot the figure
plt.figure()
plt.plot(x, f(x,a))
plt.scatter(x1, y1, color='C1', s=50)
plt.plot(xrange, line(xrange, x1, y1), 'C1--', linewidth = 2)
#optionally highlight the area between two lines - actual position vs
#predicted (if it wasn't for acceleration)
plt.fill_between(x,f(x,a),line(x, x1, y1),color='purple')
```

```python
#2
import scipy
from sympy import init_printing,lambdify
init_printing()

G,r,l,me,ms=symbols('G r l m_e m_s', real=True)
l=2.7E40
me=5.9722E24
ms=1.989E30
G=scipy.constants.G
Vr=l**2/(2*me*r**2)-G*me*ms/r
Vr
```

```python

dr=diff(Vr,r)
dr
```

```python
func=lambdify(r,Vr)
drfunc=lambdify(r,dr)
print(drfunc(2))
x=np.linspace(1E-1,1,100)
dy=[func(d) for d in x]
plt.plot(x,dy)
```
