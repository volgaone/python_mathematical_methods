```python
#3
def my_cosine(x,m,k):
    return np.exp(-m*x)*np.cos(2*np.pi*k*x)

x = np.linspace(0,1,100)
m=2
k=3
mk = [[2,3],[1,0.5],[10,15]]
y = [[my_cosine(t,m,k) for t in x] for m,k in mk]
y1 = np.transpose(y)
print(np.shape(y1))
fig, ax = plt.subplots(figsize=(4, 4))
ax.plot(x,y1)

ind = [10,20,30]
for line_num in range(np.shape(mk)[0]):
    ax.annotate(f'm{mk[line_num][0]},k{mk[line_num][1]}',
            xy=(x[ind[line_num]], y1[ind[line_num],line_num]), xycoords='data',
            xytext=(x[ind[line_num]]+0.1, y1[ind[line_num],line_num]+0.1), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='center', verticalalignment='top')
```

```python
#4
print(np.shape(y1))
print(np.shape(np.array(y1)[:,1]))
f, a = plt.subplots(3, 1, sharex=True)


i=0
colors = ["red", "black", "purple"]
styles = ["solid", "dotted", "dashed"]
widths = [1,2,3]
for ax in a:
    ax.plot(x, np.array(y1)[:,i],color = colors[i],linestyle = styles[i], linewidth=widths[i])
    # https://matplotlib.org/stable/users/explain/text/mathtext.html
    ax.set_title(r'$e^{-mx}cos(2\pi kx)$')
    ax.set_xlabel(f'm={mk[i][0]}, k={mk[i][1]}')
    i=i+1
f.tight_layout()
plt.savefig('num4ans.png')
```

```python
#5
fy2 = lambda x: pow(x,2)/10
fy3 = lambda x: np.sin(x)
x1 = np.linspace(0,6,100)
y2 = [fy2(t) for t in x1]
y3 = [fy3(t) for t in x1]
f, a = plt.subplots(3, 1, sharex=True)
a[0].plot(x1,y2)
a[1].plot(x1,y3)
a[2].plot(x1,y2,label=r"$x^2/10$")
a[2].plot(x1,y3,label=r"$sin(x)$")
f.legend(loc='lower left')
plt.savefig('num5ans.png')
```

```python
#6
from sympy import symbols, solve

xs = symbols('x', real=True)
solx = solve(xs**3+xs**2+1,xs)
for root in solx:
    print(f'{root}\n')
    number = f'{root.evalf()}'
    ntype = "complex" if "I" in number else "real"
    print(f'{number} {ntype}\n\n')
```

```python
#10
theta = np.linspace(0,2*np.pi,100)
r = [np.abs(np.cos(angle)) for angle in theta]
r1 = [np.abs(np.cos(angle)*np.sin(angle)) for angle in theta]
fig = plt.figure()
ax = fig.add_subplot(projection='polar')
ax.plot(theta,r,linewidth=1)
ax.plot(theta,r1,linewidth=1.5)
```

```python
#11
t = np.arange(0, 9, 0.1) # grid of points on x-axis
X = t
Y = 1+2*np.cos(t)
Z = -1 + 3*np.sin(t)
fig = plt.figure()
axes = fig.add_subplot(projection ='3d')
axes.plot(X, Y, Z, color='skyblue')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_zlabel('Z')
axes.set_title(" " * 35 + "Parametric line" + " " * 35)
plt.show()
```

```python
#13
def cos_approx(x,N):
    product = 1
    for k in range(1,N):
        product = product * (1- 4*x**2/(np.pi**2*(2*k-1)**2))
    return product

print(np.cos(1))
print(cos_approx(1,10))
print(cos_approx(1,50))
print(cos_approx(1,100))
```
