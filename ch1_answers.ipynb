```python
#3
def my_cosine(x,m,k):
    return np.exp(-m*x)*np.cos(2*np.pi*k*x)

x = np.linspace(0,1,100)
m=2
k=3
mk = [[2,3],[1,0.5],[10,15]]
y = [[my_cosine(t,m,k) for t in x] for m,k in mk]
y1 = np.transpose(y)
print(np.shape(y1))
fig, ax = plt.subplots(figsize=(4, 4))
ax.plot(x,y1)

ind = [10,20,30]
for line_num in range(np.shape(mk)[0]):
    ax.annotate(f'm{mk[line_num][0]},k{mk[line_num][1]}',
            xy=(x[ind[line_num]], y1[ind[line_num],line_num]), xycoords='data',
            xytext=(x[ind[line_num]]+0.1, y1[ind[line_num],line_num]+0.1), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='center', verticalalignment='top')
```

```python
#4
print(np.shape(y1))
print(np.shape(np.array(y1)[:,1]))
f, a = plt.subplots(3, 1, sharex=True)


i=0
colors = ["red", "black", "purple"]
styles = ["solid", "dotted", "dashed"]
widths = [1,2,3]
for ax in a:
    ax.plot(x, np.array(y1)[:,i],color = colors[i],linestyle = styles[i], linewidth=widths[i])
    # https://matplotlib.org/stable/users/explain/text/mathtext.html
    ax.set_title(r'$e^{-mx}cos(2\pi kx)$')
    ax.set_xlabel(f'm={mk[i][0]}, k={mk[i][1]}')
    i=i+1
f.tight_layout()
plt.savefig('num4ans.png')
```

```python
#6
from sympy import symbols, solve

xs = symbols('x', real=True)
solx = solve(xs**3+xs**2+1,xs)
for root in solx:
    print(f'{root}\n')
    number = f'{root.evalf()}'
    ntype = "complex" if "I" in number else "real"
    print(f'{number} {ntype}\n\n')
```

```python
#13
def cos_approx(x,N):
    product = 1
    for k in range(1,N):
        product = product * (1- 4*x**2/(np.pi**2*(2*k-1)**2))
    return product

print(np.cos(1))
print(cos_approx(1,10))
print(cos_approx(1,50))
print(cos_approx(1,100))
```

```python
#15
def pi_approx(N):
    sum=0
    for n in range(1,N):
        sum = sum + (-1)**(n+1)/n**2
    return sum

print(np.pi**2/12)
print(pi_approx(10))
print(pi_approx(50))
print(pi_approx(100))
```

```python
#18
from sympy import sin,pi,symbols,Sum,oo
from sympy.abc import n,x
def fourier_approx(x,N,L):
    sum=0
    for n in range(1,N):
        sum = sum + 50/np.pi*1**(n-1)/n*np.sin(n*np.pi/L*x)
    return sum

L = symbols('L', real=True)
L=5
x=1
sum = Sum(1**(n-1)/n*sin(n*pi/L*x),(n,0,oo)).doit()
print(50/pi*sum)
print(fourier_approx(1,100,5))
```

```python
#26
from sympy import diff
x=symbols('x', positive=True, real=True)
expr = x**-6-x**-12
d1=diff(expr)
d2=diff(expr,x,x)
d1
```

```python
d2
```

```python
from sympy import lambdify,solve
l_expr = lambdify(x,expr)
ld_expr = lambdify(x,d1)
x1 = np.linspace(1,5,1000)
y = [l_expr(t) for t in x1]
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(x1,y,label="f(x)")
d = [ld_expr(t) for t in x1]
ax2.plot(x1,d,color='purple',label="diff(f(x))")
print(solve(d1))
print(2**(1/6))
ax2.scatter(2**(1/6),ld_expr(2**(1/6)),color='purple',label="maxima")
fig.legend(loc='upper right')
```

```python
#27
import scipy
def Bose_Einstein(E,T):
    Kb=scipy.constants.Boltzmann
    return 1/(np.exp(E/(Kb*T))-1)

x=np.linspace(scipy.constants.electron_volt,scipy.constants.electron_volt*30,100)
fig, ax1 = plt.subplots()
temps = [2000,4000,6000]
for T in temps:
    y=[Bose_Einstein(e,T) for e in x]
    ax1.plot(x,y,')
```
