```python
from sympy import exp,pi,I,re,im
z=(exp(5*pi*I/6)+exp(-pi*I/4))/(3+I)**2
z
```

```python
r=re(z)
r
```

```python
imag=im(z)
imag
```

```python
from matplotlib import pyplot as plt
import matplotlib.ticker as ticker
ax=plt.subplot()
ax.scatter(r,imag)
ax.set_xlim([-0.1, 0.1])
ax.set_ylim([-0.1, 0.1])
#major
ax.xaxis.set_major_locator(ticker.MultipleLocator(base=0.05))
ax.yaxis.set_major_locator(ticker.MultipleLocator(base=0.05))
#minor
ax.xaxis.set_minor_locator(ticker.MultipleLocator(base=0.01))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(base=0.01))
ax.tick_params(axis='x', which='minor', bottom=False)
plt.grid(which='minor',color='r',linestyle='--')
plt.grid(which='major',color='black',linestyle='-.')
```
